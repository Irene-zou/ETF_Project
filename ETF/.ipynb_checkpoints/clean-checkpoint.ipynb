```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "../Resources2/country_vaccinations.csv"
world_vac_df = pd.read_csv(csv_file)
world_vac_df.sample(50)
```

```python
csv_file = "../Resources2/covid_impact_on_airport_traffic.csv"
ap_df = pd.read_csv(csv_file)
ap_df.sample(50)
```

```python
# Create a filtered dataframe from specific columns on World_vac
world_vac_cols = ["License Serial Number", "Premises Name", "County ID Code"]
world_vac_transformed= world_vac_df[world_vac_cols].copy()

# Rename the column headers
world_vac_transformed = world_vac_transformed.rename(columns={"License Serial Number": "id",
                                                          "Premises Name": "premise_name",
                                                          "County ID Code": "county_id"})

# Clean the data by dropping duplicates and setting the index
world_vac_transformed.drop_duplicates("id", inplace=True)
world_vac_transformed.set_index("id", inplace=True)
world_vac_transformed.head()
```

```python
## Create a filtered dataframe from specific columns on ap
ap_df_cols = ["ID", "County Name (Licensee)", "County ID Code", "License Count"]
ap_transformed = ap_df[ap_df_cols].copy()

# Rename the column headers
ap_transformed = ap_transformed.rename(columns={"ID": "id",
                                                         "County Name (Licensee)": "county_name",
                                                         "License Count": "license_count",
                                                         "County ID Code": "county_id"})

# Set index
ap_transformed.set_index("id", inplace=True)

ap_transformed.head()
```
