```python
import pandas as pd
from sqlalchemy import create_engine
import scipy.stats as sts
import numpy as np
from datetime import datetime
```

```python
#read csv file - country_vaccinations.csv
csv_file = "../Resources/country_vaccinations.csv"
world_vac_df = pd.read_csv(csv_file)
```

```python
#list vaccines provider
world_vac_df["vaccines"].value_counts()
```

```python
#clean the provider name
world_vac_df["vaccines"].replace({"Pfizer/BioNTech": "Pfizer", "Moderna, Pfizer/BioNTech ":"Moderna","Oxford/AstraZeneca, Pfizer/BioNTech ":"Vaccitech","Pfizer/BioNTech, Sinopharm":"Sinopharm", "CNBG, Sinovac":"CNBG","Oxford/AstraZeneca, Sinopharm ":"Pfizer", "Pfizer/BioNTech, Sinopharm, Sputnik V ":"Pfizer","Covaxin, Oxford/AstraZeneca":"Covaxin", "Oxford/AstraZeneca, Sinovac ":"Vaccitech","Pfizer/BioNTech, Sinovac":"Pfizer","Oxford/AstraZenec":"Vaccitech"}, inplace=True)

#extract data: relevant columns - country,iso_code, date, daily_vaccinations, vaccines,source_name
world_vac_df = world_vac_df[["country", "date", "daily_vaccinations",
                       "vaccines","source_name"]]
#rename the column
world_vac_df = world_vac_df.rename(columns = {'country': 'Country','date':'Date'})
```

```python
#vaccine provider tables
Pfizer_df = world_vac_df.loc[world_vac_df["vaccines"] == "Pfizer"]
Pfizer_df = Pfizer_df[["Country", "Date","vaccines",]]

Moderna_df = world_vac_df.loc[world_vac_df["vaccines"] == "Moderna"]
Moderna_df = Moderna_df[["Country", "Date","vaccines",]]

Vaccitech_df = world_vac_df.loc[world_vac_df["vaccines"] == "Vaccitech"]
Vaccitech_df  = Vaccitech_df[["Country", "Date", "vaccines",]]

Sinopharm_df = world_vac_df.loc[world_vac_df["vaccines"] == "Sinopharm"]
Sinopharm_df = Sinopharm_df[["Country", "Date","vaccines",]]

Sputnik_V_df = world_vac_df.loc[world_vac_df["vaccines"] == "Sputnik V"]
Sputnik_V_df = Sputnik_V_df[["Country", "Date","vaccines",]]

CNBG_df = world_vac_df.loc[world_vac_df["vaccines"] == "CNBG"]
CNBG_df = CNBG_df[["Country", "Date","vaccines",]]

Covaxin_df = world_vac_df.loc[world_vac_df["vaccines"] == "Covaxin"]
Covaxin_df = Covaxin_df[["Country", "Date",
                       "vaccines",]]
```

```python
#read impact on airport traffic csv file
csv_file = "../Resources/covid_impact_on_airport_traffic.csv"
airport_traffic_df = pd.read_csv(csv_file)
```

```python
#extract data: relevant columns - Date,Country,State,City, AirportName,PercentOfBaseline
airport_traffic_df = airport_traffic_df[["Date","Country","State","City","AirportName","PercentOfBaseline"]]
airport_traffic_df

##Clean up
 #Date format
airport_traffic_df['Date'] = airport_traffic_df['Date'].str.replace('/','-')
airport_traffic_df['Date'] = pd.to_datetime(airport_traffic_df.Date)
airport_traffic_df['Date'] = airport_traffic_df['Date'].dt.strftime('%Y-%m-%d')

 #Change countries to make it consistant across mutiple csv files
airport_traffic_df['Country'] = airport_traffic_df['Country'].replace('United States of America (the)', 'United States')

#group by date to make it easier to merge with other df and better mathching information

airport_traffic_df.head()
```
