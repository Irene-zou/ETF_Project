```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "../Resources2/country_vaccinations.csv"
world_vac_df = pd.read_csv(csv_file)
world_vac_df.sample(50)
```

```python
csv_file = "../Resources2/covid_impact_on_airport_traffic.csv"
ap_df = pd.read_csv(csv_file)
ap_df.sample(50)
```

```python
# Create a filtered dataframe from specific columns on World_vac
world_vac_cols = ["country", "date", "total_vaccinations"]
world_vac_transformed= world_vac_df[world_vac_cols].copy()

# Clean the data by dropping duplicates and setting the index
world_vac_transformed.drop_duplicates("country", inplace=True)
world_vac_transformed.set_index("country", inplace=True)
world_vac_transformed.head()
```

```python
## Create a filtered dataframe from specific columns on ap
ap_df_cols = ["Country", "Date", "AirportName", "PercentOfBaseline"]
ap_transformed = ap_df[ap_df_cols].copy()


# Set index
ap_transformed.set_index("Country", inplace=True)

ap_transformed.head()
```
